# Sequential Model - Single neuron
This notebook continues from 'MinimizingLoss.ipynb'. Basically what I did in there can all be done by TensorFlow itself. This notebook demonstrates that.

```python
import tensorflow as tf
import numpy as np

# xs is the data input, Questions
xs = [-1.0, 0.0, 1.0, 2.0, 3.0, 4.0]
# ys is the training input, Answers
ys = [-3.0, -1.0, -1.0, 3.0, 5.0, 7.0]
```

## Define & Compile the Model
Simplest model offered by TFlow is Sequential. First we define how many neural network layers this model is going to use. Here we want to use only one layer. That means we want a single neuron, the perecptron.
We are also trying to say here that the neuron is going to have a single input with a single dimension. Then we pass this layer definition to Sequential Model class and compile the model.   

```python
layer = tf.keras.layers.Dense(units=1, input_shape=(1,))
model = tf.keras.models.Sequential(layer)
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Training the model
Training the model or fitting the model can be done by using the model class.

```python
model.fit(np.array(xs, dtype=float), np.array(ys, dtype=float), epochs=500)
```

## Drawing Inference

```python
print(model.predict([10.0]))
print(model.get_weights())
```
